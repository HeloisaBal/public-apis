# Cats API

```python
import requests

cats = requests.get("https://cat-fact.herokuapp.com/facts")
print("Cats:", cats.status_code)
```

```python
# print(cats.text)
```

```python
import json

results = cats.json()
results
```

```python
import pandas as pd
pd.set_option("display.max_rows", 0)

data = pd.DataFrame(results)
data.head(10)
```

```python
data.dtypes
```

```python
pd.set_option('display.max_colwidth', None)
data['text']
```
